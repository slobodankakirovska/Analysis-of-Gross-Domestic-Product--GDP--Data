# Опис на проект:

### Бруто-домашен производ (БДП)

- Бруто-домашен производ (БДП) (англиски: gross domestic product, GDP) — еден од основните показатели на големината и успехот на една економија. Бруто-домашниот производ е дефиниран како вкупната пазарна вредност на сите произведени финални добра и услуги во една земја во одреден период (вообичаено, една година). Според тоа, во БДП се вклучуваат само финалните добра, т.е. се исклучува вредноста на интермедијарите производи во производството. Изразот бруто се однесува на фактот дека амортизацијата на капиталните фондови е вклучена во пресметката на БДП. Значи, доколку се одземе амортизацијата на капиталните фондови би се добил нето домашен производ. Бруто-домашниот производ е централна макроекономска категорија преку која се мери вкупната економска активност на една земја. Во центарот на националните сметки е производството на стоките и услугите, односно бруто-домашниот производ (БДП), кој е главен показател за развиеноста на економијата на секоја држава. Со податокот за бруто-домашниот производ по жител се врши споредба на развиеноста на државата во однос на други држави, споредба на продуктивноста на трудот и истовремено податок од кој се добиваат податоците за бруто-националниот доход, бруто-штедењето и нето -задолжувањето на државата.

Бруто-домашен производ = Финална потрошувачка + бруто-инвестиции + извоз - увоз

### Финална потрошувачка

#### Трошоци за финална потрошувачка

- Трошоците за финална потрошувачка се составени од трошоци што ги прават резидентните институционални единици за стоки и за услуги што се употребуваат за директно задоволување на личните потреби или желби, како и на колективните потреби на членовите на заедницата. Трошоците за финална потрошувачка може да се направат на домашна територија или во странство.

Финална потрошувачка= Финална потрошувачка на домаќинствата + финална потрошувачка на НПИСД + финална потрошувачка на државата.

### Бруто-инвестиции

#### ИНВЕСТИЦИИ ВО ОСНОВНИ СРЕДСТВА

Инвестициите во основни средства се мерат преку вкупната вредност на набавените минус отуѓените основни средства од производителот во текот на пресметковниот период, плус некои унапредувања на вредноста на непроизведените средства реализирани преку производната активност на институционалната единица.

### Libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---

### Барање 1:

```python
df = pd.read_csv('GDP_data.csv')
df
```

---

### Барање 2:

```python
# Extract the first 4 characters and store in a new column
df['Year'] = df['Unnamed: 0'].str[:4]

# Extract the remaining characters and store in another column
df['Quartal'] = df['Unnamed: 0'].str[4:6]

df
```

```python
df = df.drop('Unnamed: 0.1', axis=1)
df = df.drop('Unnamed: 0', axis = 1)
```

```python
# Replace 'T' with 'Q' in the 'Quartal' column
df['Quartal'] = df['Quartal'].str.replace(r'Т(\d+)(\.\d+)?', r'Q\1', regex=True)
df['Quartal'] = df['Quartal'].str.replace('T', 'Q')

# Convert specific columns to string type
columns_to_convert = ['Финална потрошувачка на домаќинствата, вклучувајќи НПИСД',
                      'Финална јавна потрошувачка', 'Бруто-инвестиции',
                      'Извоз на стоки и на услуги', 'Увоз на стоки и на услуги']
df[columns_to_convert] = df[columns_to_convert].astype(str)

# Remove whitespace and convert specific columns to float type
df[columns_to_convert] = df[columns_to_convert].apply(lambda x: x.str.replace(' ', '')).astype(float)

# Convert 'Year' column to int type
df['Year'] = df['Year'].astype(int)

# Check the data types of the columns
print(df.dtypes)
df
```

---

### Барање 3:

```python
df['Бруто домашен производ'] = df['Финална јавна потрошувачка'] + df['Бруто-инвестиции'] + df['Извоз на стоки и на услуги'] - df['Увоз на стоки и на услуги']
df
```

---

### Барање 4:

```python
filt_years = [2020, 2021, 2022]
total_gdp = df[(df['Year'].isin(filt_years)) & (df['Quartal'].isin(['Q1', 'Q2']))].groupby('Year')['Бруто домашен производ'].sum()
#print(total_gdp)

#Plotting the pie chart

plt.figure(figsize=(8, 6))
plt.pie(total_gdp, labels=total_gdp.index, autopct='%1.1f%%')
plt.title('Bкупната вредност на БДП од првите два квартали за 2020, 2021 и 2022 година')
plt.show()

# Vrednosta na BDP bila najvisoka za prvite dva kvartala vo 2022, a za okolu 10% pomalku vo 2020 godina.
```

### Барање 5:

```python
desired_years = [2020, 2021, 2022]
desired_quarters = ['Q1', 'Q2', 'Q3', 'Q4']
filtered_data = df[df['Year'].isin(desired_years) & df['Quartal'].isin(desired_quarters)]

# Create the boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Year', y='Бруто домашен производ', hue='Quartal', data=filtered_data)
plt.title('Boxplot of GDP for 2020, 2021, and 2022 by Quarters')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Quarter')
plt.show()
```

```python
# BDP za prviot kvartal vo 2020 i prviot kvartal vo 2021 e recisi identicen, vtor kvartal e znacitelno pomal vo 2020 za razlika od 2021.
# Tretiot i cetvrtiot kvartal sledat nekoj rast, eden vo odnos na drug. Za 2022 godina imame podatoci samo za prvite dva kvartala
# koi od prethodnite 2 godini se so pointenziven rast na BDP.
```

### Барање 6:

```python
filtered_data = df[df['Year'].isin(df['Year'].unique())]

# Create the scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(filtered_data['Year'], filtered_data['Бруто домашен производ'])
plt.title('Scatter Plot of GDP for All Years')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

```python
# Vrednostite sledat trend na rast, se gleda od grafikot, kako vrednostite rechisi pravilno se rasporedeni vo edna nagorna linija.
```

### Барање 7:

```python
# Calculate the total GDP for each year
yearly_gdp = df.groupby('Year')['Бруто домашен производ'].sum()

# Find the years where the total GDP is above 60000
years_above_60000 = yearly_gdp[yearly_gdp > 60000].index.tolist()

print("Years with total GDP above 60000:", years_above_60000)
```

```python
# Create the bar plot
plt.bar(yearly_gdp.index, yearly_gdp)

# Set the x-axis labels with rotation
plt.xticks(yearly_gdp.index, rotation=45)

# Set the y-axis label
plt.ylabel('БДП вредност ')

# Set the title
plt.title('Total GDP by Year')

# Add a horizontal line at y = 60000
plt.axhline(y=60000, color='r', linestyle='--')

# Show the plot
plt.show()

# Spored podatocite sto se doneseni na obrabotka i godinite sto se opservirani, kaj site godini BDP e nad 60000.
```

### Барање 8:

```python
yearly_consumption = df.groupby('Year')['Финална јавна потрошувачка'].sum()

print("Финална јавна потрошувачка за секоја година:")
print(yearly_consumption)
```

```python
# Create a pivot table to reshape the data for the heatmap
pivot_table = df.pivot_table(index='Year', values='Финална јавна потрошувачка', aggfunc='sum')

# Create the heatmap using seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=False, cmap="YlGnBu")
plt.title("Вкупна финална потрошувачка за секоја година")
plt.ylabel("Година")
plt.show()

```
